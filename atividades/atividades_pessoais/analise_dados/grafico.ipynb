### Criar grafico com uma base de dados csv

```python
# passo 1 : importar as bibliotecas e a base de dados 

import pandas as pd
import plotly.express as px

tabela = pd.read_csv('arquivo.csv')
```

```python
# passo 2 : analisar e remover dados que nao são uteis 
display(tabela)

# osdados inuteis : id_cliente ,  o id não influencia na satisfação do cliente 

tabela.drop(columns='id_cliente')
display(tabela)
```

```python
# passo 3 : criar os graficos 
for coluna in tabela.columns:

    grafico = px.histogram(tabela, x=coluna, color='status_churn')
    # churn é uma metrica de porcentagem de clientes que deixaram de usar os produtos, ou servisos , de uma empresa em um determinado periodo

    # exibir o grafico

    grafico.show() 
```
