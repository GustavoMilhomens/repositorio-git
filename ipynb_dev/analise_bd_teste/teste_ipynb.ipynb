```python
# passo 1: garantir a existencia do arquivo garantindo a existencia de cada tipo de dado 
import os

tipo_dados = "id_cliente,num_cliente,massa_bolo,recheio_bolo1,recheio_bolo2,data_pedido,tipo_bolo"

if os.path.exists('teste.csv') == False:
    with open('base_dados.csv','w') as file: # se não existir o arquivo no diretorio o sistema cria um 
        file.write(tipo_dados)
```

```python
# passo 2: importar a base de dados 
import pandas as pd

# esse dicionario é para ter menos chances de perder os dados iniciais


n_cliente = input('numero do cliente: ')
massa_bolo = input('massa do bolo: ')
recheio_bolo1 = input('primeiro recheio')
recheio_bolo2 = input('segundo recheio ')
data_bolo = input('data de entrega')
tipo = input('tipo do bolo')
dados = "\n" + n_cliente + ',' + massa_bolo  + ',' + recheio_bolo1  + ',' + recheio_bolo2  + ',' + data_bolo  + ',' + tipo
# newline serve para evitar linhas em branco extras
# encoding='utf-8' é uma boa prática para lidar com caracteres especiais.
# o 'a' vem de append que serve para escrever ouadicionar novos dados 
with open('teste.csv','a', newline='',encoding='utf-8') as file:
    file.write(dados) 


texto = pd.read_csv('teste.csv')
display(texto)
```

```python
import os
import csv

dados = [['ID', 'Nome'], [1, 'Alice'], [2, 'Bob']]
nome_arquivo = 'dados_salvos.csv'

try:
    with open(nome_arquivo, 'w', newline='', encoding='utf-8') as arquivo_csv:
        writer = csv.writer(arquivo_csv)
        writer.writerows(dados)
    print(f"O arquivo '{nome_arquivo}' foi fechado corretamente pelo 'with' e salvo.")

    # Verificação da existência do arquivo
    if os.path.exists(nome_arquivo):
        print(f"Confirmação: O arquivo '{nome_arquivo}' existe no diretório.")
    else:
        print(f"Atenção: O arquivo '{nome_arquivo}' NÃO foi encontrado após a escrita.")

except IOError as e:
    print(f"Erro ao escrever no arquivo: {e}")
```
